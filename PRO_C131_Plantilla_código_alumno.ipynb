# Explorando sistemas solares con estadísticas.

En la clase de hoy, vamos a trazar diferentes gráficas y aplicaremos estadísticas a nuestros datos para ver si podemos encontrar alguna información interesante sobre diferentes sistemas solares

\
Antes de eso, recordemos el significado de todas las columnas que quedan en nuestro CSV -

* ***name*** - *(nombre)*: el nombre del planeta.
* ***light_years_from_earth*** - *(años luz de la tierra)*: Distancia del exoplaneta a la tierra en años luz.
* ***planet_mass*** - *(planeta_masa)*: masa del planeta.
* ***stellar_magnitude*** - *(magnitud estelar)*: este es el brillo de la estrella anfitriona del planeta cuando se observa desde la Tierra (al igual que el sol es nuestra estrella anfitriona).
* ***discovery_date*** - *(fecha de descubrimiento)*: este es el año del descubrimiento del exoplaneta.
planet_type - (planeta_tipo): este es el tipo de planeta (gigante gaseoso, supertierra, etc.).
* ***planet_radius*** - *(planeta_radio)*: este es el radio del exoplaneta con respecto a la Tierra o Júpiter.
* ***orbital_radius*** - *(radio de órbita)*: esta es la distancia promedio de este exoplaneta a su sol. Al igual que nuestro sistema solar tiene 1 sol, varios sistemas solares contienen muchos planetas y soles.
orbital_period - (período de órbita): este es el tiempo que tarda en completar una órbita de su sol.
* ***eccentricity*** - *(excentricidad)*: esto denota cuán circular es la órbita. También puede tener forma ovalada. Cuanto menor es la excentricidad, más circular es la órbita.
* ***solar_system_name*** - *(nombre del sistema solar)*: el nombre del sistema solar anfitrión.
* ***planet_discovery_method*** - *(método de descubrimiento del planeta)*: este es el método de descubrimiento que se utilizó para encontrar este exoplaneta.
* ***planet_orbital_inclination*** - *(inclinación orbital del planeta)*: esta es la inclinación orbital, lo que significa que es la inclinación de la órbita del exoplaneta cuando gira alrededor de su sol.
* ***planet_density*** - *(densidad del planeta)*: esta es la densidad del planeta.
right_ascension - (ascensión recta): esta es la ascensión recta del sistema planetario, que es la coordenada este-oeste por la cual se mide la posición de este planeta.
* ***declination*** - *(declinación)*: esta es la coordenada norte-sur por la cual se mide la posición del planeta.
* ***host_temperature*** - *(anfitriona_temperatura)*: esta es la temperatura de la estrella anfitriona en Kelvin.
* ***host_mass*** - *(anfitriona_masa)*: esta es la cantidad de masa contenida en la estrella anfitriona.
* ***host_radius*** - *(anfitriona_radio)*: este es el radio de la estrella anfitriona.

Ahora, escribimos el código para leer el CSV.

```
import csv

rows = []

with open("/content/PRO-NASA-Exoplanet-Processed-Data/main.csv", "r") as f:
  csvreader = csv.reader(f)
  for row in csvreader:
    rows.append(row)

headers = rows[0]
planet_data_rows = rows[1:]
print(headers)
print(planet_data_rows[0])
```

Aquí, podemos ver que se agregó un campo adicional al CSV, que indica el conteo de filas, pero el primer encabezado es una cadena vacía. ¡Arreglemos esto y luego procedamos a encontrar la cantidad de planetas en cada uno de los sistemas solares!

```
headers[0] = "row_num"

solar_system_planet_count = {}
for planet_data in planet_data_rows:
  if solar_system_planet_count.get(planet_data[11]):
    solar_system_planet_count[planet_data[11]] += 1
  else:
    solar_system_planet_count[planet_data[11]] = 1

max_solar_system = max(solar_system_planet_count, key=solar_system_planet_count.get)
print("¡El sistema solar {} tiene el máximo número de planetas {} de todos los sistemas solares que hemos descubierto hasta ahora!".format(max_solar_system, solar_system_planet_count[max_solar_system]))
```

¡Aquí podemos ver que hay un sistema solar conocido como **KOI-351** donde hay **8** planetas! ¡¡Eso es genial!!
¿Podría ser esta nuestra próxima casa? ¡Obtengamos la lista de todos los planetas de este sistema solar, para poderlos ver más de cerca!

```
temp_planet_data_rows = list(planet_data_rows)
for planet_data in temp_planet_data_rows:
  planet_mass = planet_data[3]
  if planet_mass.lower() == "unknown":
    planet_data_rows.remove(planet_data)
    continue
  else:
    planet_mass_value = planet_mass.split(" ")[0]
    planet_mass_ref = planet_mass.split(" ")[1]
    if planet_mass_ref == "Jupiters":
      planet_mass_value = float(planet_mass_value) * 317.8
    planet_data[3] = planet_mass_value

  planet_radius = planet_data[7]
  if planet_radius.lower() == "unknown":
    planet_data_rows.remove(planet_data)
    continue
  else:
    planet_radius_value = planet_radius.split(" ")[0]
    planet_radius_ref = planet_radius.split(" ")[2]
    if planet_radius_ref == "Jupiter":
      planet_radius_value = float(planet_radius_value) * 11.2
    planet_data[7] = planet_radius_value

print(len(planet_data_rows))

koi_351_planets = []
for planet_data in planet_data_rows:
  if max_solar_system == planet_data[11]:
    koi_351_planets.append(planet_data)

print(len(koi_351_planets))
print(koi_351_planets)
```

Genial, ahora tracemos una gráfica de barras con la columna **planet_mass** - *(masa del planeta)* con los datos de estos 8 planetas.

```
import plotly.express as px

koi_351_planet_masses = []
koi_351_planet_names = []
for planet_data in koi_351_planets:
  koi_351_planet_masses.append(planet_data[3])
  koi_351_planet_names.append(planet_data[1])

koi_351_planet_masses.append(1)
koi_351_planet_names.append("Tierra")

fig = px.bar(x=koi_351_planet_names, y=koi_351_planet_masses)
fig.show()
```

Ahora, antes de que continuemos, aprendamos un dato curioso.

El gran científico Albert Einstein nos dio una ecuación con la que podemos calcular la gravedad de cualquier planeta.
La ecuación es esta:

$$g = \frac{G*M_{Tierra}}{d^2}$$

Aquí, $G$ es una constante gravitacional, lo que significa que siempre será la misma.

$M_{Tierra}$ es la masa de la Tierra (o cualquier otro planeta si lo estamos calculando para otro planeta).

$d$ es el radio del planeta.

Aquí, podemos ver una relación inversa entre el radio del planeta y la gravedad. Cuanto mayor sea el radio (y más grande el planeta), menor será la gravedad.

Pero entonces, también vemos una relación directa entre la masa del planeta y la gravedad. Cuanto mayor sea la masa del planeta, mayor será la gravedad.

La gravedad de nuestra **Tierra es de  $9.8 m/s^2$** y nosotros como humanos estamos acostumbrados a ella.


Para que podamos existir en cualquier otro planeta, la gravedad debe ser cercana a la que tenemos aquí.

**Marte** tiene una gravedad de **3,711 m/s** y la **Luna** tiene una gravedad de **1,62 m/s**.


Con lo que acabamos de aprender, intentemos trazar un diagrama de dispersión para todos los planetas, donde mantendremos la masa del planeta como la coordenada Y, el radio del planeta como la coordenada X, y el tamaño del círculo como la gravedad de la misma. ¡A ver si encontramos algo interesante!

**El valor de G (constante gravitacional) es 6.674e-11**.

**Masa de la Tierra = 5.972e+24**.

**Radio de la Tierra = 6371000**.

Nota: como tenemos **planet_mass** - *(masa del planeta)* y **planet_radius** *(radio del planeta)* con referencia a la Tierra, no olvides multiplicar la masa de la Tierra y el radio de la Tierra con estos valores.

```
# Gráfica plot
planet_masses = []
planet_radiuses = []
planet_name = []

for planet_data in planet_data_rows:
  planet_masses.append(planet_data[3])
  planet_radiuses.append(planet_data[7])
  planet_name.append(planet_data[1])

planet_gravity = []

for index, name in enumerate(planet_name):
  gravity = (float(planet_masses[index])*5.972e+24)/(float(planet_radiuses[index])* float(planet_radiuses[index])*6371000)*6.674e-11
  planet_gravity.append(gravity)


```

```
fig = px.scatter(x=planet_radiuses,y=planet_masses, size=planet_gravity, hover_data=[planet_name])
fig.show()
```

**Dato curioso: ¡nuestros cuerpos humanos de pie pueden soportar una fuerza gravitacional 90 veces más fuerte que la Tierra!**

Aunque eso va a ser un poco extremo, aún podemos sobrevivir con una gravedad 10 veces mayor que la que tenemos en la Tierra. ¡Enlistemos los nombres de los planetas con gravedad igual o menor a 100!


```
#Planetas de gravedad baja.

low_gravity_planet = []

for index, gravity in enumerate(planet_gravity):
  if gravity < 100:
    low_gravity_planet.append(planet_data_rows[index])

print(len(low_gravity_planet))
```
